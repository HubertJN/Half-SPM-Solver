```python
import pybamm
```

```python
model = pybamm.lithium_ion.SPM()
```

```python
parameter_values = pybamm.ParameterValues('Chen2020')
```

```python
sim = pybamm.Simulation(model, parameter_values=parameter_values)
sim.solve([0,3600])
sim.plot()
```

```python
model.variables.search('mol.m-3')
```

```python
sim.plot(['Average positive particle concentration [mol.m-3]', 'Positive particle concentration [mol.m-3]'])
```
